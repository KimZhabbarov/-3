Гистограммы распределения и ящики с усами для всех признаков

```python
features = [
    'anxiety_level', 'self_esteem', 'mental_health_history', 'depression',
       'headache', 'blood_pressure', 'sleep_quality', 'breathing_problem',
       'noise_level', 'living_conditions', 'safety', 'basic_needs',
       'academic_performance', 'study_load', 'teacher_student_relationship',
       'future_career_concerns', 'social_support', 'peer_pressure',
       'extracurricular_activities', 'bullying'
]

def plot_visualizations(df, feature):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[feature])
    plt.title(f'Ящик с усами {feature}')

    plt.subplot(1, 2, 1)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Гистограмма {feature}')

    plt.tight_layout()
    plt.show()

for feature in features:
    plot_visualizations(data, feature)
```

Групповая визуализация

Корелляционная матрица(взаимосвязь между всеми признаками с целевой переменной - stress_level)

```python
corr_matrix = data.iloc[:,:-1].corr()
sns.heatmap(corr_matrix)
corr = abs(data.corr())['stress_level'].sort_values(ascending=False)
print(corr)
```

Парные диаграммы

```python
correlation_matrix = data.corr()
correlation_value = correlation_matrix.loc['future_career_concerns', 'bullying']
print(f"Коэффициент корреляции между 'future_career_concerns' и 'bullying': {correlation_value:.3f}")

correlation_value = correlation_matrix.loc['self_esteem', 'bullying']
print(f"Коэффициент корреляции между 'self_esteem' и 'bullying': {correlation_value:.3f}")

correlation_value = correlation_matrix.loc['future_career_concerns', 'self_esteem']
print(f"Коэффициент корреляции между 'future_career_concerns' и 'self_esteem': {correlation_value:.3f}")
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(data[['self_esteem', 'bullying', 'future_career_concerns', 'stress_level']], hue='stress_level', palette='coolwarm')
plt.show()
```

Графики показывают, что точки с разными уровнями стресса имеют разные распределения в пространстве признаков.

Графики оказывают сложные взаимосвязи между тремя переменными: future_career_concerns, bullying, и self_esteem. Несмотря на то, что диаграммы сами по себе не демонстрируют явных линейных зависимостей (точки сильно рассеяны), коэффициенты корреляции указывают на значительные связи.

Сравнение показателей в зависимости от уровня стресса

```python
plt.figure(figsize=(14, 8))

plt.subplot(1, 3, 1)
sns.boxplot(x='stress_level', y='bullying', data=data)
plt.title('Bullying по уровню стресса')

plt.subplot(1, 3, 2)
sns.boxplot(x='stress_level', y='future_career_concerns', data=data)
plt.title('Future_career_concerns по уровню стресса')

plt.subplot(1, 3, 3)
sns.boxplot(x='stress_level', y='self_esteem', data=data)
plt.title('Self_esteem сна по уровню стресса')

plt.tight_layout()
plt.show()
```

Наличие выбросов указывает на то, что некоторые люди более восприимчивы к влиянию стресса, чем другие.

Точечная диаграмма для визуализации взаимосвязи между парой c сильной корреляцией

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='future_career_concerns', y='bullying', hue='stress_level', data=data, palette='coolwarm')
plt.title('Взаимосвязь между future_career_concerns и bullying')
plt.show()
```

Гистограмма распределения самооценки в зависимости от уровня стресса

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='self_esteem', hue='stress_level', multiple='stack')
plt.axvline(data['self_esteem'].mean(), color='r', linestyle='--', label='Среднее значение')
plt.axvline(data['self_esteem'].median(), color='g', linestyle='-', label='Медиана')
plt.title('Гистограмма распределения самооценка по уровню стресса')
plt.legend()
plt.show()
```
